```python
from sklearn.datasets import load_wine
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import numpy as np


#load_wine()関数を使用してデータを取得
dataset = load_wine()
# dataset.data #数値のデータが入っている
#dataset.feature_names #要素名が入っている

#pandasのデータフレームにデータを格納して見やすくする
df = pd.DataFrame(data=dataset.data,columns=dataset.feature_names)
df.head()

#ワインの種別
# dataset.target

#データフレームにワインの種別も追加
df['category'] = dataset.target
df.head()

#レコード数とデータの列数を表示
df.shape

#データの分割の前に
#説明変数（機械学習の予測モデルにインプットする側のデータ）
#目的変数（予測値として出力される側のデータ）
#に分ける

x = dataset.data
y = dataset.target

#データを分割(テストデータを3割に設定)
#説明変数、目的変数それぞれに学習データとテストデータが割り当てされる
#random_stateの数値はどういう意味なのかを質問

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=5)
# 以下のようにrandom_stateを指定しない場合、実行毎に結果が異なります。
# train_test_split(X, y, test_size=0.3)

# print(x.shape,x_train.shape,x_test.shape,y.shape,y_train.shape,y_test.shape)

#予測モデルのインスタンス化
model = RandomForestClassifier(random_state=3)
# 以下のようにrandom_stateを指定しない場合、最終的に生成される予測モデルが実行毎に異なる可能性があります。
# これは、モデルの評価結果や予測結果が実行毎に変わってしまう可能性があること意味します。
# model = DecisionTreeClassifier()

#予測モデルの学習
model.fit(x_train,y_train)

#予測モデルの評価
y_pred = model.predict(x_test)

#accuracy_scoreを使用した評価
accuracy_score(y_test,y_pred)

#scoreを使用した評価
model.score(x_test,y_test)



```
